# Data Viz 
Heatmaps to show data sparsity across Counties/columns

```python
# Group Dataframe columns into lists based on what the data represents
weather = ['CloudCover', 'Other Conditions', 'PrecipType', 'Rainfall', 'Rainfall Within 24 Hours',
          'Rainfall Within 48 Hours', 'Weather Conditions Day Before Yesterday', 'Weather Conditions Today', 
          'Weather Conditions Yesterday', 'WindDirection', 'WindSpeed']
water = ['Sea State', 'Stream Flow', 'Tidal Stage', 'TideStage', 'Water Color', 'Water Color Description', 'Water Odor',
        'Water Surfaces', 'WaveHeight']
measure = ['Parameter', 'Lab', 'ParameterName_CBP', 'ParameterName_CMC', 'SampleType', 'Unit']
admin = ['Agency', 'Cruise', 'Date', 'Source', 'Station', 'state', 'HUCNAME_']
```

```python
# Create a list that will contain missing column data for each column Group

#county_tpl = []
#measure_tpl = []
water_tpl = []
```

```python
# Create a list of unique counties
county_ls = df['COUNTY_'].unique()

# For each county, loop through each column in the desired column group (e.g. 'weather') and find proportion of
# rows FILLED OUT... higher number means more data!
for county in county_ls:
    for col in water:
        perc_filled = (df[df['COUNTY_'] == county][col].count() / len(df[df['COUNTY_'] == county]))
        
        # Add the % of column filled out for this county and column to the dataset
        water_tpl.append((county, col, perc_filled))
```

```python
# Create a DataFrame for County / Column / % Filled for the Weather column Group
weather_df = pd.DataFrame(county_tpl, columns = ['County', 'Column', 'Percent Filled'])
```

```python
measure_df = pd.DataFrame(measure_tpl, columns = ['County', 'Column', 'Percent Filled'])
```

```python
water_df = pd.DataFrame(water_tpl, columns = ['County', 'Column', 'Percent Filled'])
```

```python
weather_df.head()
```

```python
# Filter County/Column data so it must be a number that's at least 0 (no null)
weather_df2 = weather_df[weather_df['Percent Filled'] >= 0]
```

```python
# Create a dataset to create heatmap from
heatmap1_data = pd.pivot_table(weather_df2, values='Percent Filled', 
                     index=['Column'], 
                     columns='County')
```

```python
# Set the dimensions of the heatmap (width, height)
fig, ax = plt.subplots(figsize=(15,3))

# Plot the heatmap for the desired data and set the color scheme
sns.heatmap(heatmap1_data, cmap="YlGnBu")
```
